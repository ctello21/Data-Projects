# Combine all of the code above into one function

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python

def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    crypto_name = soup.find('span', class_ = 'sc-f70bb44c-0 jltoa', title = 'Bitcoin').text
    crypto_name = title_split[0]

    crypto_price = soup.find('span', class_ = 'sc-f70bb44c-0 jxpCgO base-text').text

    final_price = crypto_price.replace('$','')

    date_time = datetime.now()

    dict = {'Crypto Name':crypto_name,
            'Price':final_price,
            'TimeStamp':date_time}

    df = pd.DataFrame([dict])

    if os.path.exists(r'C:\Users\isctl\OneDrive - IRI\Documents\~TRAINING~\AnalystBuilder\Python - Automated Web Scraper Project\Crypto_Automated_Pull.csv'):
        df.to_csv(r'C:\Users\isctl\OneDrive - IRI\Documents\~TRAINING~\AnalystBuilder\Python - Automated Web Scraper Project\Crypto_Automated_Pull.csv', mode='a', header = False, index = False)
    else:
        df.to_csv(r'C:\Users\isctl\OneDrive - IRI\Documents\~TRAINING~\AnalystBuilder\Python - Automated Web Scraper Project\Crypto_Automated_Pull.csv', index = False)
    print(df)
```

### Runs automated_crypto_pull after a set time and inputs data into CSV

```python
while True:
    automated_crypto_pull()
    time.sleep(10)
```
